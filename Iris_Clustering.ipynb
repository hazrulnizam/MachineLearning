```
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = load_iris()

X = data['data']

sepallength = X[:,0]
#sepallength = sepallength.reshape((sepallength.shape[0], 1))

sepalwidth = X[:,1]
#sepalwidth = sepalwidth.reshape((sepalwidth.shape[0], 1))

petallength = X[:,2]
#petallength = petallength.reshape((petallength.shape[0], 1))

petalwidth = X[:,3]
#petalwidth = petalwidth.reshape((petalwidth.shape[0], 1))

fig, axes = plt.subplots(2, 2)
axes[0,0].scatter(x=sepallength, y=sepallength)
axes[0,0].set_xlabel('Sepal Length (cm)')
axes[0,0].set_ylabel('Sepal Length (cm)')

axes[0,1].scatter(x=sepallength, y=sepalwidth)
axes[0,1].set_xlabel('Sepal Length (cm)')
axes[0,1].set_ylabel('Sepal Width (cm)')

axes[1,0].scatter(x=sepallength, y=petallength)
axes[1,0].set_xlabel('Sepal Length (cm)')
axes[1,0].set_ylabel('Petal Length (cm)')

axes[1,1].scatter(x=sepallength, y=petalwidth)
axes[1,1].set_xlabel('Sepal Length (cm)')
axes[1,1].set_ylabel('Petal Width (cm)')

plt.show()
```

```
from sklearn.cluster import KMeans

Data = {'x': petallength,
        'y': petalwidth
       }
  
df = pd.DataFrame(Data,columns=['x','y'])
  
kmeans = KMeans(n_clusters=2).fit(df)
y_pred = kmeans.predict(df)
centroids = kmeans.cluster_centers_
# print(centroids)

# plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(df['x'], df['y'], c=y_pred)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
```

```
from sklearn.cluster import KMeans

Data = {'x': petallength,
        'y': petalwidth,
        'v': sepallength,
        'w': sepalwidth
       }
  
df = pd.DataFrame(Data,columns=['x','y','v','w'])
  
kmeans = KMeans(n_clusters=3).fit(df)
y_pred = kmeans.predict(df)
centroids = kmeans.cluster_centers_
# print(centroids)

# plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(df['w'], df['y'], c=y_pred)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
```
