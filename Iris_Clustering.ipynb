```
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = load_iris()

X = data['data']

sepallength = X[:,0]
#sepallength = sepallength.reshape((sepallength.shape[0], 1))

sepalwidth = X[:,1]
#sepalwidth = sepalwidth.reshape((sepalwidth.shape[0], 1))

petallength = X[:,2]
#petallength = petallength.reshape((petallength.shape[0], 1))

petalwidth = X[:,3]
#petalwidth = petalwidth.reshape((petalwidth.shape[0], 1))

plt.scatter(x=sepallength, y=sepalwidth)
plt.title('Scatter Plot of Sepal Width vs Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```
plt.scatter(x=sepallength, y=petallength)
plt.title('Scatter Plot of Petal Length vs Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```
plt.scatter(x=petallength, y=petalwidth)
plt.title('Scatter Plot of Petal Width vs Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```
from sklearn.cluster import KMeans

Data = {'x': petallength,
        'y': petalwidth
       }
  
df = pd.DataFrame(Data,columns=['x','y'])
  
kmeans = KMeans(n_clusters=2).fit(df)
y_pred = kmeans.predict(df)
centroids = kmeans.cluster_centers_
# print(centroids)

# plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(df['x'], df['y'], c=y_pred)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
```
